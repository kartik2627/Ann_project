```python
import streamlit as st
from transformers import AutoModelForCausalLM, AutoTokenizer
import torch
```

```python
MODEL_NAME = "microsoft/DialoGPT-medium"  # Using an open-access model
tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME)
model = AutoModelForCausalLM.from_pretrained(MODEL_NAME)
```

```python
def chatbot_response(user_input, intent):
    """Generates a response based on user input and selected intent."""
    prompt = f"User intent: {intent}\nUser: {user_input}\nChatbot:"  # Context formatting
    inputs = tokenizer(prompt, return_tensors="pt")
    outputs = model.generate(**inputs, max_length=200)
    response = tokenizer.decode(outputs[0], skip_special_tokens=True)
    return response.split("Chatbot:")[-1].strip()
```

```python
MODEL_NAME = "microsoft/DialoGPT-medium"  # Using an open-access model
tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME)
model = AutoModelForCausalLM.from_pretrained(MODEL_NAME)

def chatbot_response(user_input, intent):
    """Generates a response based on user input and selected intent."""
    prompt = f"User intent: {intent}\nUser: {user_input}\nChatbot:"  # Context formatting
    inputs = tokenizer(prompt, return_tensors="pt")
    with torch.no_grad():  # Disable gradient calculation for inference
        outputs = model.generate(**inputs, max_length=200)
    response = tokenizer.decode(outputs[0], skip_special_tokens=True)
    return response.split("Chatbot:")[-1].strip()

# Streamlit UI
st.title("🤖 Advanced AI Chatbot")
st.markdown("Select an intent and start chatting!")

user_input = st.text_input("Type your message here...")
intent = st.radio("Choose Intent", [
    "General Conversation",
    "Technical Support",
    "Health Advice",
    "Travel Guide"
])

if st.button("Send"):
    if user_input.strip():  # Ensure the input is not empty
        response = chatbot_response(user_input, intent)
        st.text_area("Chatbot Response", response, height=200)
    else:
        st.warning("Please enter a message before sending.")
```
